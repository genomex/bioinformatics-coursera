```python
def Skew(Genome):
    skew = [0]
    for i in range(0, len(Genome)):
        if Genome[i] == 'C':
            skew.append(skew[i] - 1)
        elif Genome[i] == 'G':
            skew.append(skew[i] + 1)
        else:
            skew.append(skew[i])
    return skew
```

```python
def Skew_explained(Genome, verbose):
    
    
    skew_list = [0]
    current_skew = 0
    
    
    for i in range(len(Genome)):
        
        if Genome[i] == 'C':
            skew_list.append(skew_list[i] - 1)
            current_skew -= 1
            
            
            
        elif Genome[i] == 'G':
            skew_list.append(skew_list[i] + 1)
            current_skew += 1
            
            
        else:
            skew_list.append(skew_list[i])
            
        if verbose:
            print(Genome[i],current_skew)
            
            
    return skew_list
```

```python
def min_skew(Genome):
    
    skew = Skew(Genome)
    
    minimum = min(skew)
    
    return [i for i, val in enumerate(skew) if val == minimum]
```
